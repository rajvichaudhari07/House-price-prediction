I've chosen MAPE and r2 metrics to rate the model because of easy interpretation. I'll be using this metrics for all models in this work.

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, r2_score
```

```python
# create error function 

def error(y_true,y_pred):
  print('MAE: ', mean_absolute_error(y_true,y_pred))
  print('MAPE: ', mean_absolute_percentage_error(y_true,y_pred))
  print('r2 score: ', r2_score(y_true,y_pred))
```

```python
# model's forecasting accuracy 

error(test['price'],test['price_pred'])
```

#Let's check, does renovation affect the price. Tramsform 'renovation_year' in binary categorical feature, where 0 is building without renovation and 1 is building after renovation in any year.

```python
def renovation_cat(column):
    if column == 0:
        return 0
    else:
        return 1
```

```python
df
```

```python
# create new column for that feature

df['renovation'] = df['renovation_year'].apply(renovation_cat)
```

```python
df['renovation']
```

```python
# count buildings in both classes

df.renovation.value_counts()
```

```python
sns.boxplot(data=df, x='renovation', y='price', palette='Dark2_r')
plt.title('Relate of the price and renovation');
```

It's hard to interpret relation with outliers. Let's build relation plot of interquartile range of the prices and renovation.

```python
# Select the first quartile
q1 = df['price'].quantile(.25)

# Select the third quartile
q3 = df['price'].quantile(.75)

# Create a mask beetween q1 & q3
mask = df['price'].between(q1, q3, inclusive='both')

# Filtering the initial dataframe with a mask
iqr_of_price = df.loc[mask, 'price']
```
